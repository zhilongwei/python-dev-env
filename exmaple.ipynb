```python
import numpy as np
import sympy
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rc

# Configure MathText to mimic LaTeX fonts (Computer Modern)
mpl.rcParams['text.usetex'] = False               # Make sure we're *not* using external LaTeX
mpl.rcParams['mathtext.fontset'] = 'cm'           # Use the Computer Modern font set
mpl.rcParams['mathtext.rm'] = 'Computer Modern'   # Roman
mpl.rcParams['mathtext.it'] = 'Computer Modern:italic'
mpl.rcParams['mathtext.bf'] = 'Computer Modern:bold'

# Optional: set a default, serif-based font family for non-math text as well
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 10
```

```python
plt.figure(figsize=(6, 4))
plt.plot([1, 2, 3], [4, 5, 6], label=r"$y = x + 3$")
plt.xlabel(r"$t$ (s)")
plt.ylabel(r"$u$ (m/s)")
plt.show()
```
